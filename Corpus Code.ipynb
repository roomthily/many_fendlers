Quick data processing for the corpus.



```python
%reload_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/vernaculars.csv', header=0, names=['TSN', 'Taxon1', 'Taxon2', 'Vernacular'])
```

```python
df[:10]
```

```python
ndf = df.dropna()
ndf[:10]
```

```python
# we're ignoring the warning today
ndf['init_token'] = ndf['Vernacular'].str.split().str[0]

```

```python
ndf
```

```python
ndf['check_token'] = ndf['Vernacular'].str.split().str[1:].str.join(' ')
```

```python
ndf[:25]
```

```python
ndf['check_token'].replace('', np.nan, inplace=True)
fdf = ndf.dropna()
```

```python
fdf[:25]
```

```python
unique_checks = fdf['check_token'].unique().tolist()
len(unique_checks)
```

```python
with open('data/check_tokens.txt', 'w') as f:
    f.write('\n'.join(unique_checks))
```
